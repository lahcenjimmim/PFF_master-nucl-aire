```python
import os
import pandas as pd

print("Lancement de l'extraction des données")

#  PRÉPARATION
mes_donnees = {}
liste_dossiers = ["Pb_DD-ME2", "Pb_DD-PC1"]

# LECTURE DES FICHIERS
for dossier in liste_dossiers:
    if os.path.exists(dossier):
        for nom_fichier in os.listdir(dossier):
            if nom_fichier.endswith(".out"):
                chemin_complet = os.path.join(dossier, nom_fichier)
                
                
                nom_1 = nom_fichier.replace(".out", "")
                morceaux = nom_1.split("_")
                
                A = None
                
                for morceau in morceaux:
                    if morceau.isdigit(): 
                        A = int(morceau)
                        break 
             
                
                if A is not None:
                    
                    if A not in mes_donnees:
                        mes_donnees[A] = {"Isotope Pb": A}
                    
                    # On cherche l'énergie dans le fichier
                    with open(chemin_complet, "r") as fichier:
                        for ligne in fichier:
                            if "total" in ligne.lower() and "energy" in ligne.lower():
                                # On isole l'énergie et on convertit
                                valeur_texte = ligne.split()[-1].replace("D", "e").replace("d", "e")
                                energie = abs(float(valeur_texte))
                                
                                # On ajoute l'énergie dans la bonne colonne
                                nom_modele = dossier.replace("Pb_", "")
                                mes_donnees[A][nom_modele] = energie
                                break 

#  AFFICHAGE DU TABLEAU
print("\n Voici les résultats :")

if mes_donnees:
    # On transforme en liste et on trie par Isotope
    liste_finale = sorted(list(mes_donnees.values()), key=lambda x: x["Isotope Pb"])
    tableau = pd.DataFrame(liste_finale)
    display(tableau)
else:
    print(" aucune donnée n'a été extraite.")
```

```python
import matplotlib.pyplot as plt

print("Création du graphique : Énergie de liaison par nucléon (E/A)")

# Le Plomb (Pb) a  82 protons
Z_Pb = 82

# ==========================================
#  PRÉPARATION DES DONNÉES THÉORIQUES 
# ==========================================
axe_X_N_ME2 = []
axe_Y_BEA_ME2 = []

axe_X_N_PC1 = []
axe_Y_BEA_PC1 = []


if 'mes_donnees' in locals():
   
    isotopes_tries = sorted(mes_donnees.keys())

   
    for A in isotopes_tries:
        # Calcul du nombre de neutrons
        N = A - Z_Pb
        
        
        if "DD-ME2" in mes_donnees[A]:
            energie_totale = abs(mes_donnees[A]["DD-ME2"])
            energie_par_nucleon = energie_totale / A
            axe_X_N_ME2.append(N)
            axe_Y_BEA_ME2.append(energie_par_nucleon)
            
        
        if "DD-PC1" in mes_donnees[A]:
            energie_totale = abs(mes_donnees[A]["DD-PC1"])
            energie_par_nucleon = energie_totale / A
            axe_X_N_PC1.append(N)
            axe_Y_BEA_PC1.append(energie_par_nucleon)

# ==========================================
#  PRÉPARATION DES DONNÉES EXPÉRIMENTALES
# ==========================================

A_exp = [178, 180, 182, 184, 186, 188, 190, 192, 194, 196, 198, 200, 202, 204, 206, 208, 210, 212, 214, 216, 218, 220]

# Énergies par nucléon E/A correspondantes
EA_exp = [7.69084, 7.72570, 7.75633, 7.78273, 7.80534, 7.82482, 7.84113, 7.85465, 
          7.86542, 7.87334, 7.87897, 7.88181, 7.88215, 7.87993, 7.87536, 7.86745, 
          7.83596, 7.80432, 7.77239, 7.740, 7.705, 7.670]

# Calcul du nombre de neutrons pour les données expérimentales
axe_X_N_Exp = [A - Z_Pb for A in A_exp]
axe_Y_BEA_Exp = EA_exp

# ==========================================
#  PRÉPARATION DES DONNÉES FRDM-2012 (Isotopes pair-pair)
# ==========================================
# Les neutrons (N) extraits de la table
axe_X_N_FRDM = [96, 98, 100, 102, 104, 106, 108, 110, 112, 114, 116, 118, 120, 122, 124, 126, 128, 130, 132, 134, 136, 138, 140, 142, 144, 146, 148, 150, 152, 154, 156, 158, 160, 162, 164, 166, 168, 170, 172, 174, 176, 178, 180, 182, 184, 186, 188, 190]

# L'énergie de liaison TOTALE (E_bind) extraite
E_bind_FRDM = [1368.62, 1390.33, 1411.43, 1431.91, 1451.82, 1471.17, 1490.01, 1508.35, 1526.18, 1543.48, 1560.26, 1576.49, 1592.21, 1607.40, 1622.08, 1635.60, 1645.39, 1654.18, 1662.65, 1670.84, 1678.91, 1686.70, 1694.20, 1701.62, 1709.33, 1716.57, 1723.58, 1730.34, 1736.52, 1741.90, 1746.90, 1751.88, 1756.81, 1761.44, 1765.39, 1769.51, 1773.22, 1776.98, 1780.18, 1782.91, 1785.54, 1787.61, 1789.65, 1791.34, 1792.29, 1790.77, 1788.63, 1787.10]


# Transformation de l'énergie de liaison totale en Énergie par nucléon (E/A)
axe_Y_BEA_FRDM = []
for i in range(len(axe_X_N_FRDM)):
    N = axe_X_N_FRDM[i]
    A = N + Z_Pb
    E_tot = E_bind_FRDM[i]
    energie_par_nucleon = E_tot / A
    axe_Y_BEA_FRDM.append(energie_par_nucleon)


# ==========================================
#  LE DESSIN DU GRAPHIQUE
# ==========================================

plt.figure(figsize=(10, 6)) 


if axe_X_N_ME2: 
    plt.plot(axe_X_N_ME2, axe_Y_BEA_ME2, color="blue", marker="o", linestyle="-", label="DIRHB: DD-ME2")


if axe_X_N_PC1:
    plt.plot(axe_X_N_PC1, axe_Y_BEA_PC1, color="red", marker="s", linestyle="-", label="DIRHB: DD-PC1")


plt.plot(axe_X_N_FRDM, axe_Y_BEA_FRDM, color="black", marker="d", linestyle="--", label="Modèle FRDM (2012)")


plt.plot(axe_X_N_Exp, axe_Y_BEA_Exp, color="green", marker="^", markersize=8, linestyle="", label="Données Expérimentales")


plt.title("Énergie de liaison par nucléon pour les isotopes de Plomb (Pb pair-pair)")
plt.xlabel("Nombre de neutrons (N)")
plt.ylabel("Énergie par nucléon E/A (MeV)")


plt.legend(loc="lower center")      
plt.grid(True, linestyle=":", alpha=0.7)   
plt.savefig('energie_Pb_complet.png', dpi=300, bbox_inches='tight')

# On affiche le résultat final 
plt.show()
```

```python
import matplotlib.pyplot as plt

print("Calcul et tracé de l'énergie de séparation à deux neutrons (S2n)")

Z_Pb = 82

# ==========================================
# PRÉPARATION DES DONNÉES THÉORIQUES (DIRHB)
# ==========================================
axe_X_N_ME2 = []
axe_Y_S2n_ME2 = []

axe_X_N_PC1 = []
axe_Y_S2n_PC1 = []

if 'mes_donnees' in locals():
   
    isotopes_tries = sorted(mes_donnees.keys())

    # CALCUL DE S2n POUR LES MODÈLES THÉORIQUES
    for A in isotopes_tries:
        
        if A % 2 != 0:
            continue
            
        N = A - Z_Pb
        A_moins_2 = A - 2 
        
        # Modèle DD-ME2
        if "DD-ME2" in mes_donnees[A] and A_moins_2 in mes_donnees and "DD-ME2" in mes_donnees[A_moins_2]:
            s2n = abs(mes_donnees[A]["DD-ME2"]) - abs(mes_donnees[A_moins_2]["DD-ME2"])
            axe_X_N_ME2.append(N)
            axe_Y_S2n_ME2.append(s2n)
            
        # Modèle DD-PC1 
        if "DD-PC1" in mes_donnees[A] and A_moins_2 in mes_donnees and "DD-PC1" in mes_donnees[A_moins_2]:
            s2n = abs(mes_donnees[A]["DD-PC1"]) - abs(mes_donnees[A_moins_2]["DD-PC1"])
            axe_X_N_PC1.append(N)
            axe_Y_S2n_PC1.append(s2n)

# ==========================================
#  PRÉPARATION DES DONNÉES EXPÉRIMENTALES
# ==========================================
A_exp = [178, 180, 182, 184, 186, 188, 190, 192, 194, 196, 198, 200, 202, 204, 206, 208, 210, 212, 214, 216, 218, 220]
EA_exp = [7.69084, 7.72570, 7.75633, 7.78273, 7.80534, 7.82482, 7.84113, 7.85465, 
          7.86542, 7.87334, 7.87897, 7.88181, 7.88215, 7.87993, 7.87536, 7.86745, 
          7.83596, 7.80432, 7.77239, 7.740, 7.705, 7.670]

axe_X_N_Exp = []
axe_Y_S2n_Exp = []


dict_energie_totale_exp = {}
for i in range(len(A_exp)):
    dict_energie_totale_exp[A_exp[i]] = A_exp[i] * EA_exp[i]

# Calcul du S2n expérimental
for A in A_exp:
    A_moins_2 = A - 2
    if A_moins_2 in dict_energie_totale_exp:
        s2n = dict_energie_totale_exp[A] - dict_energie_totale_exp[A_moins_2]
        axe_X_N_Exp.append(A - Z_Pb)
        axe_Y_S2n_Exp.append(s2n)

# ==========================================
# PRÉPARATION DES DONNÉES FRDM-2012
# ==========================================
N_FRDM = [96, 98, 100, 102, 104, 106, 108, 110, 112, 114, 116, 118, 120, 122, 124, 126, 128, 130, 132, 134, 136, 138, 140, 142, 144, 146, 148, 150, 152, 154, 156, 158, 160, 162, 164, 166, 168, 170, 172, 174, 176, 178, 180, 182, 184, 186, 188, 190]
E_bind_FRDM = [1368.62, 1390.33, 1411.43, 1431.91, 1451.82, 1471.17, 1490.01, 1508.35, 1526.18, 1543.48, 1560.26, 1576.49, 1592.21, 1607.40, 1622.08, 1635.60, 1645.39, 1654.18, 1662.65, 1670.84, 1678.91, 1686.70, 1694.20, 1701.62, 1709.33, 1716.57, 1723.58, 1730.34, 1736.52, 1741.90, 1746.90, 1751.88, 1756.81, 1761.44, 1765.39, 1769.51, 1773.22, 1776.98, 1780.18, 1782.91, 1785.54, 1787.61, 1789.65, 1791.34, 1792.29, 1790.77, 1788.63, 1787.10]

axe_X_N_FRDM_S2n = []
axe_Y_S2n_FRDM = []

dict_energie_totale_frdm = dict(zip(N_FRDM, E_bind_FRDM))

for N in N_FRDM:
    N_moins_2 = N - 2
    if N_moins_2 in dict_energie_totale_frdm:
        s2n = dict_energie_totale_frdm[N] - dict_energie_totale_frdm[N_moins_2]
        axe_X_N_FRDM_S2n.append(N)
        axe_Y_S2n_FRDM.append(s2n)

# ==========================================
#  LE DESSIN DU GRAPHIQUE
# ==========================================
plt.figure(figsize=(10, 6))

if axe_X_N_ME2:
    plt.plot(axe_X_N_ME2, axe_Y_S2n_ME2, color="blue", marker="o", label="DIRHB: DD-ME2")
if axe_X_N_PC1:
    plt.plot(axe_X_N_PC1, axe_Y_S2n_PC1, color="red", marker="s", linestyle="--", label="DIRHB: DD-PC1")

plt.plot(axe_X_N_FRDM_S2n, axe_Y_S2n_FRDM, color="black", marker="^", linestyle="-.", label="Modèle FRDM (2012)")

if axe_X_N_Exp:
    plt.plot(axe_X_N_Exp, axe_Y_S2n_Exp, color="green", marker="d", markersize=8, linestyle="", label="Données Expérimentales")


axes = plt.gca()
ymin, ymax = axes.get_ylim()


plt.ylim(bottom=ymin, top=ymax * 1.1)


hauteur_texte = ymax * 1.05 

plt.axvline(x=126, color="gray", linestyle=":")
plt.text(126.5, hauteur_texte, "N=126\n(Fermeture de couche)", color="gray", fontweight="bold", verticalalignment='top')

plt.axvline(x=184, color="gray", linestyle=":")
plt.text(184.5, hauteur_texte, "N=184\n(Fermeture de couche)", color="gray", fontweight="bold", verticalalignment='top')

plt.title("Énergie de séparation à deux neutrons ($S_{2n}$) pour les isotopes de Plomb")
plt.xlabel("Nombre de neutrons (N)")
plt.ylabel("Énergie $S_{2n}$ (MeV)")


plt.legend(loc="lower left")
plt.grid(True, linestyle=":", alpha=0.7)
plt.savefig('Energie_separation_S2n_Pb_Corrigé.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
import matplotlib.pyplot as plt

print("Calcul et tracé de delta_2n (Signature des fermetures de couches)")

Z_Pb = 82

# ==========================================
#  PRÉPARATION DES DONNÉES THÉORIQUES (DIRHB)
# ==========================================
axe_X_N_ME2, axe_Y_D2n_ME2 = [], []
axe_X_N_PC1, axe_Y_D2n_PC1 = [], []

if 'mes_donnees' in locals():
    
    isotopes_tries = sorted(mes_donnees.keys())

    # CALCUL DE DELTA_2n
    # Formule : Delta_2n(N) = S2n(N) - S2n(N+2)
    for A in isotopes_tries:
        N = A - Z_Pb
        A_prec = A - 2
        A_suiv = A + 2
        
       
        # Pour le modèle DD-ME2
        if all(x in mes_donnees and "DD-ME2" in mes_donnees[x] for x in [A_prec, A, A_suiv]):
            s2n_actuel = abs(mes_donnees[A]["DD-ME2"]) - abs(mes_donnees[A_prec]["DD-ME2"])
            s2n_suivant = abs(mes_donnees[A_suiv]["DD-ME2"]) - abs(mes_donnees[A]["DD-ME2"])
            delta2n = s2n_actuel - s2n_suivant
            
            axe_X_N_ME2.append(N)
            axe_Y_D2n_ME2.append(delta2n)

        # Pour le modèle DD-PC1
        if all(x in mes_donnees and "DD-PC1" in mes_donnees[x] for x in [A_prec, A, A_suiv]):
            s2n_actuel = abs(mes_donnees[A]["DD-PC1"]) - abs(mes_donnees[A_prec]["DD-PC1"])
            s2n_suivant = abs(mes_donnees[A_suiv]["DD-PC1"]) - abs(mes_donnees[A]["DD-PC1"])
            delta2n = s2n_actuel - s2n_suivant
            
            axe_X_N_PC1.append(N)
            axe_Y_D2n_PC1.append(delta2n)

# ==========================================
# PRÉPARATION DES DONNÉES EXPÉRIMENTALES
# ==========================================
A_exp = [178, 180, 182, 184, 186, 188, 190, 192, 194, 196, 198, 200, 202, 204, 206, 208, 210, 212, 214, 216, 218, 220]
EA_exp = [7.69084, 7.72570, 7.75633, 7.78273, 7.80534, 7.82482, 7.84113, 7.85465, 
          7.86542, 7.87334, 7.87897, 7.88181, 7.88215, 7.87993, 7.87536, 7.86745, 
          7.83596, 7.80432, 7.77239, 7.740, 7.705, 7.670]


dict_energie_totale_exp = {A_exp[i]: A_exp[i] * EA_exp[i] for i in range(len(A_exp))}

axe_X_N_Exp, axe_Y_D2n_Exp = [], []

# Calcul du Delta_2n expérimental
for A in A_exp:
    A_prec = A - 2
    A_suiv = A + 2
    
    if A_prec in dict_energie_totale_exp and A_suiv in dict_energie_totale_exp:
        s2n_actuel = dict_energie_totale_exp[A] - dict_energie_totale_exp[A_prec]
        s2n_suivant = dict_energie_totale_exp[A_suiv] - dict_energie_totale_exp[A]
        delta2n = s2n_actuel - s2n_suivant
        
        axe_X_N_Exp.append(A - Z_Pb)
        axe_Y_D2n_Exp.append(delta2n)


# ==========================================
# PRÉPARATION DES DONNÉES FRDM-2012
# ==========================================
N_FRDM = [96, 98, 100, 102, 104, 106, 108, 110, 112, 114, 116, 118, 120, 122, 124, 126, 128, 130, 132, 134, 136, 138, 140, 142, 144, 146, 148, 150, 152, 154, 156, 158, 160, 162, 164, 166, 168, 170, 172, 174, 176, 178, 180, 182, 184, 186, 188, 190]
E_bind_FRDM = [1368.62, 1390.33, 1411.43, 1431.91, 1451.82, 1471.17, 1490.01, 1508.35, 1526.18, 1543.48, 1560.26, 1576.49, 1592.21, 1607.40, 1622.08, 1635.60, 1645.39, 1654.18, 1662.65, 1670.84, 1678.91, 1686.70, 1694.20, 1701.62, 1709.33, 1716.57, 1723.58, 1730.34, 1736.52, 1741.90, 1746.90, 1751.88, 1756.81, 1761.44, 1765.39, 1769.51, 1773.22, 1776.98, 1780.18, 1782.91, 1785.54, 1787.61, 1789.65, 1791.34, 1792.29, 1790.77, 1788.63, 1787.10]


dict_energie_totale_frdm = dict(zip(N_FRDM, E_bind_FRDM))

axe_X_N_FRDM_D2n = []
axe_Y_D2n_FRDM = []

for N in N_FRDM:
    N_prec = N - 2
    N_suiv = N + 2
    
    
    if N_prec in dict_energie_totale_frdm and N_suiv in dict_energie_totale_frdm:
        s2n_actuel = dict_energie_totale_frdm[N] - dict_energie_totale_frdm[N_prec]
        s2n_suivant = dict_energie_totale_frdm[N_suiv] - dict_energie_totale_frdm[N]
        delta2n = s2n_actuel - s2n_suivant
        
        axe_X_N_FRDM_D2n.append(N)
        axe_Y_D2n_FRDM.append(delta2n)


# ==========================================
#  LE DESSIN DU GRAPHIQUE
# ==========================================
plt.figure(figsize=(10, 6))

if axe_X_N_ME2:
    plt.plot(axe_X_N_ME2, axe_Y_D2n_ME2, color="blue", marker="o", label="DIRHB: DD-ME2")
if axe_X_N_PC1:
    plt.plot(axe_X_N_PC1, axe_Y_D2n_PC1, color="red", marker="s", linestyle="--", label="DIRHB: DD-PC1")

# On trace la courbe FRDM-2012 (en violet)
plt.plot(axe_X_N_FRDM_D2n, axe_Y_D2n_FRDM, color="black", marker="d", linestyle="-.", label="Modèle FRDM (2012)")

# On trace la courbe expérimentale
if axe_X_N_Exp:
    plt.plot(axe_X_N_Exp, axe_Y_D2n_Exp, color="green", marker="^", markersize=8, linestyle="", label="Données Expérimentales")


axes = plt.gca()
ymin, ymax = axes.get_ylim()


plt.ylim(bottom=ymin, top=ymax * 1.15)
hauteur_texte = ymax * 1.05 

# Lignes pour les nombres magiques (Pics attendus) 
plt.axvline(x=126, color="gray", linestyle=":")
plt.axvline(x=184, color="gray", linestyle=":")


plt.text(126.5, hauteur_texte, "N=126\n(Fermeture de couche)", color="gray", fontweight="bold", verticalalignment='top')
plt.text(184.5, hauteur_texte, "N=184\n(Fermeture de couche)", color="gray", fontweight="bold", verticalalignment='top')

plt.title("Signature de la fermeture en couches ($\Delta_{2n}$) - Plomb (Pb)")
plt.xlabel("Nombre de neutrons (N)")
plt.ylabel("$\Delta_{2n}$ (MeV)")


plt.legend(loc="upper left")
plt.grid(True, linestyle=":", alpha=0.6)
plt.savefig('Signature_de_la_fermeture_en_couches_D2n.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
import os
import matplotlib.pyplot as plt

print("Extraction et tracé de l'énergie d'appariement (Pairing)")

Z_Pb = 82
liste_dossiers = ["Pb_DD-ME2", "Pb_DD-PC1"]

# EXTRACTION DES DONNÉES
donnees_pairing = {}

for dossier in liste_dossiers:
    if os.path.exists(dossier):
        for nom_fichier in os.listdir(dossier):
            if nom_fichier.endswith(".out"):
                chemin_complet = os.path.join(dossier, nom_fichier)
                
               
                A = None
                for morceau in nom_fichier.replace(".out", "").split("_"):
                    if morceau.isdigit():
                        A = int(morceau)
                        break
                
                if A is not None:
                    if A not in donnees_pairing:
                        donnees_pairing[A] = {}
                        
                    e_pair = None
                    nom_modele = dossier.replace("Pb_", "")
                    
                   
                    with open(chemin_complet, "r") as fichier:
                        for ligne in fichier:
                            ligne_min = ligne.lower()
                            if "e_pair" in ligne_min or "e-pair" in ligne_min or "pairing energy" in ligne_min:
                                valeur_texte = ligne.split()[-1].replace("D", "e").replace("d", "e")
                                e_pair = abs(float(valeur_texte))
                                
                   
                    if e_pair is not None:
                        donnees_pairing[A][nom_modele] = e_pair

# PRÉPARATION DES LISTES POUR LE TRACÉ
axe_X_N_ME2, axe_Y_Pair_ME2 = [], []
axe_X_N_PC1, axe_Y_Pair_PC1 = [], []


isotopes_tries = sorted(donnees_pairing.keys())

for A in isotopes_tries:
    N = A - Z_Pb
    if "DD-ME2" in donnees_pairing[A]:
        axe_X_N_ME2.append(N)
        axe_Y_Pair_ME2.append(donnees_pairing[A]["DD-ME2"])
        
    if "DD-PC1" in donnees_pairing[A]:
        axe_X_N_PC1.append(N)
        axe_Y_Pair_PC1.append(donnees_pairing[A]["DD-PC1"])

# LE DESSIN DU GRAPHIQUE
plt.figure(figsize=(10, 6))

plt.plot(axe_X_N_ME2, axe_Y_Pair_ME2, color="blue", marker="o", label="DD-ME2")
plt.plot(axe_X_N_PC1, axe_Y_Pair_PC1, color="red", marker="s", linestyle="--", label="DD-PC1")



plt.axhline(y=0, color="black", linestyle="-")


plt.axvline(x=126, color="gray", linestyle=":")
plt.text(126.5, plt.ylim()[1] * 0.9, "N=126", color="gray", fontweight="bold")

plt.axvline(x=184, color="gray", linestyle=":")
plt.text(184.5, plt.ylim()[1] * 0.9, "N=184", color="gray", fontweight="bold")


plt.title("Énergie d'appariement neutronique pour le Plomb (Pb)")
plt.xlabel("Nombre de neutrons (N)")
plt.ylabel("Énergie d'appariement |E_pair| (MeV)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig('Énergie d''appariement.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import os
import pandas as pd

print("Lancement de l'extraction des rayons de charge")

# PRÉPARATION
mes_donnees_rayon = {}
liste_dossiers = ["Pb_DD-ME2", "Pb_DD-PC1"]

#  LECTURE DES FICHIERS
for dossier in liste_dossiers:
    if os.path.exists(dossier):
        for nom_fichier in os.listdir(dossier):
            if nom_fichier.endswith(".out"):
                chemin_complet = os.path.join(dossier, nom_fichier)
                
               
                nom_sans_extension = nom_fichier.replace(".out", "")
                morceaux = nom_sans_extension.split("_")
                
                A = None
                for morceau in morceaux:
                    if morceau.isdigit(): 
                        A = int(morceau)
                        break 
                # -----------------------------
                
                if A is not None:
                    
                    if A not in mes_donnees_rayon:
                        mes_donnees_rayon[A] = {"Isotope Pb": A}
                    
                    rayon_charge = None
                    nom_modele = dossier.replace("Pb_", "")
                    
                    
                    with open(chemin_complet, "r") as fichier:
                        for ligne in fichier:
                            ligne_min = ligne.lower()
                            
                            
                            if "charge" in ligne_min and "radius" in ligne_min:
                                
                                
                               
                                try:
                                    valeur_texte = ligne.split()[-1].replace("D", "e").replace("d", "e")
                                    rayon_charge = abs(float(valeur_texte))
                                except ValueError:
                                    
                                    valeur_texte = ligne.split()[-2].replace("D", "e").replace("d", "e")
                                    rayon_charge = abs(float(valeur_texte))
                                    
                               

                    
                    if rayon_charge is not None:
                        mes_donnees_rayon[A][nom_modele] = rayon_charge

# AFFICHAGE DU TABLEAU
print("\n Voici les résultats :")

if mes_donnees_rayon:
    # On transforme en liste et on trie par Isotope
    liste_finale = sorted(list(mes_donnees_rayon.values()), key=lambda x: x["Isotope Pb"])
    tableau = pd.DataFrame(liste_finale)
    display(tableau)
else:
    print(" aucune donnée n'a été extraite.")
```

```python
import os
import pandas as pd

print("Lancement de l'extraction des énergies de liaison pour le Mercure (Hg)")

#  PRÉPARATION
mes_donnees_hg = {}
liste_dossiers = ["Hg_DD-ME2", "Hg_DD-PC1"] 
Z_Hg = 80

#  LECTURE DES FICHIERS
for dossier in liste_dossiers:
    if os.path.exists(dossier):
        for nom_fichier in os.listdir(dossier):
            if nom_fichier.endswith(".out"):
                chemin_complet = os.path.join(dossier, nom_fichier)
                
              
                nom_sans_extension = nom_fichier.replace(".out", "")
                morceaux = nom_sans_extension.split("_")
                
                A = None
                for morceau in morceaux:
                    if morceau.isdigit(): 
                        A = int(morceau)
                        break 
                # -----------------------------
                
                if A is not None:
                   
                    if A not in mes_donnees_hg:
                        
                        mes_donnees_hg[A] = {"Isotope Hg": A, "Neutrons N": A - Z_Hg}
                    
                    energie = None
                    nom_modele = dossier.replace("Hg_", "")
                    
                    
                    with open(chemin_complet, "r") as fichier:
                        for ligne in fichier:
                            ligne_min = ligne.lower()
                            
                            if "total" in ligne_min and "energy" in ligne_min:
                                valeur_texte = ligne.split()[-1].replace("D", "e").replace("d", "e")
                                energie = abs(float(valeur_texte))
                                break 

                    
                    if energie is not None:
                        mes_donnees_hg[A][nom_modele] = energie

#  AFFICHAGE DU TABLEAU
print("\nVoici les résultats :")

if mes_donnees_hg:
    
    liste_finale = sorted(list(mes_donnees_hg.values()), key=lambda x: x["Isotope Hg"])
    
    tableau_hg = pd.DataFrame(liste_finale)
    display(tableau_hg)
else:
    print(" aucune donnée n'a été extraite. Vérifiez que les dossiers Hg_DD-ME2 et Hg_DD-PC1 existent.")
```

```python
import matplotlib.pyplot as plt

print("Calcul et tracé de l'énergie de désintégration Alpha (Q_alpha)")

Z_Pb = 82

# On utilise la valeur expérimentale précise de l'énergie de liaison de l'Hélium-4
BE_alpha = 28.296 

# ==========================================
#  PRÉPARATION DES DONNÉES THÉORIQUES (DIRHB)
# ==========================================
axe_X_N_ME2, axe_Y_Qa_ME2 = [], []
axe_X_N_PC1, axe_Y_Qa_PC1 = [], []

if 'mes_donnees' in locals() and 'mes_donnees_hg' in locals():
    isotopes_pb_tries = sorted(mes_donnees.keys())

    # CALCUL DE Q_alpha THÉORIQUE
    for A_Pb in isotopes_pb_tries:
        N_Pb = A_Pb - Z_Pb
        A_Hg = A_Pb - 4 
        
        # Modèle DD-ME2 
        if A_Pb in mes_donnees and "DD-ME2" in mes_donnees[A_Pb] and \
           A_Hg in mes_donnees_hg and "DD-ME2" in mes_donnees_hg[A_Hg]:
            
            be_pb = mes_donnees[A_Pb]["DD-ME2"]
            be_hg = mes_donnees_hg[A_Hg]["DD-ME2"]
            
            # Application de la formule : Q = BE(Pb) - BE(Hg) - BE(alpha)
            q_alpha = be_pb - be_hg - BE_alpha
            
            axe_X_N_ME2.append(N_Pb)
            axe_Y_Qa_ME2.append(q_alpha)

        # Modèle DD-PC1 
        if A_Pb in mes_donnees and "DD-PC1" in mes_donnees[A_Pb] and \
           A_Hg in mes_donnees_hg and "DD-PC1" in mes_donnees_hg[A_Hg]:
            
            be_pb = mes_donnees[A_Pb]["DD-PC1"]
            be_hg = mes_donnees_hg[A_Hg]["DD-PC1"]
            
            q_alpha = be_pb - be_hg - BE_alpha
            
            axe_X_N_PC1.append(N_Pb)
            axe_Y_Qa_PC1.append(q_alpha)

# ==========================================
#  PRÉPARATION DES DONNÉES EXPÉRIMENTALES
# ==========================================
donnees_exp_brutes = {
    178: -7.789, 180: -7.419, 182: -7.066, 184: -6.774, 186: -6.471,
    188: -6.109, 190: -5.698, 192: -5.222, 194: -4.738, 196: -4.238,
    198: -3.692, 200: -3.110, 202: -2.589, 204: -1.968, 206: -1.135,
    208: -0.516, 210: -3.792, 212: -3.290, 214: -2.690, 216: -2.070,
    218: -1.430, 220: -0.790
}

axe_X_N_Exp = []
axe_Y_Qa_Exp = []


for A_exp in sorted(donnees_exp_brutes.keys()):
    N_exp = A_exp - Z_Pb
    q_alpha_exp = donnees_exp_brutes[A_exp]
    
    axe_X_N_Exp.append(N_exp)
    axe_Y_Qa_Exp.append(q_alpha_exp)


# ==========================================
#  PRÉPARATION DES DONNÉES FRDM-2012
# ==========================================
axe_X_N_FRDM_Qa = []
axe_Y_Qa_FRDM = []


donnees_FRDM_Pb = {
    96: 1368.62, 98: 1390.33, 100: 1411.43, 102: 1431.91, 104: 1451.82, 
    106: 1471.17, 108: 1490.01, 110: 1508.35, 112: 1526.18, 114: 1543.48, 
    116: 1560.26, 118: 1576.49, 120: 1592.21, 122: 1607.40, 124: 1622.08, 
    126: 1635.60, 128: 1645.39, 130: 1654.18, 132: 1662.65, 134: 1670.84, 
    136: 1678.91, 138: 1686.70, 140: 1694.20, 142: 1701.62, 144: 1709.33, 
    146: 1716.57, 148: 1723.58, 150: 1730.34, 152: 1736.52, 154: 1741.90, 
    156: 1746.90, 158: 1751.88, 160: 1756.81, 162: 1761.44, 164: 1765.39, 
    166: 1769.51, 168: 1773.22, 170: 1776.98, 172: 1780.18, 174: 1782.91, 
    176: 1785.54, 178: 1787.61, 180: 1789.65, 182: 1791.34, 184: 1792.29, 
    186: 1790.77
}

donnees_FRDM_Hg = {
    96: 1369.28, 98: 1389.95, 100: 1410.05, 102: 1429.67, 104: 1448.73, 
    106: 1467.18, 108: 1485.00, 110: 1502.28, 112: 1518.99, 114: 1535.21, 
    116: 1550.99, 118: 1566.23, 120: 1580.91, 122: 1595.12, 124: 1608.48, 
    126: 1620.99, 128: 1629.88, 130: 1637.79, 132: 1645.41, 134: 1652.85, 
    136: 1659.96, 138: 1667.10, 140: 1673.92, 142: 1680.97, 144: 1687.74, 
    146: 1694.25, 148: 1700.58, 150: 1706.63, 152: 1712.06, 154: 1716.76, 
    156: 1721.14, 158: 1725.54, 160: 1729.75, 162: 1733.66, 164: 1736.81, 
    166: 1739.62, 168: 1742.84, 170: 1745.82, 172: 1748.44, 174: 1750.66, 
    176: 1752.72, 178: 1754.21, 180: 1754.94, 182: 1755.49, 184: 1755.88, 
    186: 1753.73
}

# Calcul de Q_alpha pour FRDM
for N_Pb in sorted(donnees_FRDM_Pb.keys()):
    N_Hg = N_Pb - 2 
    
    
    if N_Hg in donnees_FRDM_Hg:
        be_pb = donnees_FRDM_Pb[N_Pb]
        be_hg = donnees_FRDM_Hg[N_Hg]
        
        # Formule exacte : Q = E_bind(Pb) - E_bind(Hg) - E_bind(Alpha)
        q_alpha = be_pb - be_hg - BE_alpha
        
        axe_X_N_FRDM_Qa.append(N_Pb)
        axe_Y_Qa_FRDM.append(q_alpha)


# ==========================================
# LE DESSIN DU GRAPHIQUE
# ==========================================
plt.figure(figsize=(10, 6))

if axe_X_N_ME2:
    plt.plot(axe_X_N_ME2, axe_Y_Qa_ME2, color="blue", marker="o", label="DIRHB: DD-ME2")
if axe_X_N_PC1:
    plt.plot(axe_X_N_PC1, axe_Y_Qa_PC1, color="red", marker="s", linestyle="--", label="DIRHB: DD-PC1")

# On trace la courbe FRDM-2012 (en noir)
if axe_X_N_FRDM_Qa:
    plt.plot(axe_X_N_FRDM_Qa, axe_Y_Qa_FRDM, color="black", marker="d", linestyle="-.", label="Modèle FRDM (2012)")

# On trace la courbe expérimentale
if axe_X_N_Exp:
    plt.plot(axe_X_N_Exp, axe_Y_Qa_Exp, color="green", marker="^", linestyle="", markersize=8, label="Expérimental")

# Ligne Q=0 : au-dessus de cette ligne, la désintégration alpha est possible spontanément
plt.axhline(y=0, color="black", linestyle="-", linewidth=1.5)


axes = plt.gca()
ymin, ymax = axes.get_ylim()

plt.ylim(bottom=ymin, top=ymax * 1.1)

hauteur_texte = ymax * 0.95

# Repères N=126 et N=184 (pics attendus dans le graphe de Q_alpha) 
plt.axvline(x=126, color="gray", linestyle=":")
plt.text(126.5, hauteur_texte, "N=126\n(Fermeture de couche)", color="gray", fontweight="bold")

plt.axvline(x=184, color="gray", linestyle=":")
plt.text(184.5, hauteur_texte, "N=184\n(Fermeture de couche)", color="gray", fontweight="bold")

plt.title(r"Énergie de désintégration $\alpha$ ($Q_\alpha$) pour le Plomb (Pb)")
plt.xlabel("Nombre de neutrons du noyau père (N)")
plt.ylabel(r"$Q_\alpha$ (MeV)")


plt.legend(loc="lower right")
plt.grid(True, linestyle=":", alpha=0.6)


plt.savefig('energie_de_desintegration.png', dpi=300, bbox_inches='tight')
plt.show()
```
