```python
import csv
from pypdf import PdfReader

fichier_pdf = "frdm2012.dat.pdf"
fichier_sortie = "energie_liaison_plomb_correct.csv"

print("🔍 Extraction infaillible des énergies de liaison (E_bind) pour le Plomb...")

donnees = [["Z", "A", "N", "E_bind (MeV)"]]
reader = PdfReader(fichier_pdf)
z_actuel = 0

for page in reader.pages:
    texte = page.extract_text()
    if not texte:
        continue
        
    for ligne in texte.split("\n"):
        colonnes = ligne.split()
        
        # 1. Détecter l'en-tête de la section (ex: "Z = 82 (Pb)")
        if "Z" in colonnes and "=" in colonnes:
            try:
                idx_z = colonnes.index("Z")
                z_actuel = int(colonnes[idx_z + 2])
            except:
                pass
                
        # 2. Si on est sur une ligne de données du Plomb
        elif z_actuel == 82 and len(colonnes) >= 10:
            try:
                N = int(colonnes[0])
                A = int(colonnes[1])
                
                # On filtre les noyaux entre N=96 et N=190
                if 96 <= N <= 190:
                    e_bind_correct = None
                    
                    # ASTUCE : On cherche la seule valeur > 1000 dans la ligne
                    for col in colonnes:
                        try:
                            # Remplacement du tiret spécial PDF par un vrai signe moins Python
                            val = float(col.replace('−', '-'))
                            
                            # L'énergie de liaison du Plomb est toujours > 1000 MeV
                            if val > 1000:
                                e_bind_correct = col
                                break # On a trouvé, on arrête de chercher sur cette ligne
                        except ValueError:
                            continue # Si ce n'est pas un nombre, on passe à la colonne suivante
                    
                    if e_bind_correct:
                        donnees.append([82, A, N, e_bind_correct])
                        print(f"Isotope Pb-{A} (N={N}) : E_bind = {e_bind_correct} MeV")
                        
            except ValueError:
                continue

# 3. Sauvegarde propre dans le fichier CSV
with open(fichier_sortie, "w", newline="", encoding="utf-8") as f_out:
    writer = csv.writer(f_out)
    writer.writerows(donnees)

print("-" * 50)
print(f"✅ Extraction parfaite ! Les vrais résultats sont dans '{fichier_sortie}'.")
```
